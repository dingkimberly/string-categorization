```python
import pandas as pd

data = pd.read_csv('search_strings.csv', encoding = "ISO-8859-1")
data['item_title'] = data['item_title'].str.lower()
```

```python
items = data[['item_title']]
categories = data[['category']]
```

```python
import string

def tokenize(text):
    text = text.translate(str.maketrans({key: None for key in string.punctuation}))
    text = text.translate(str.maketrans({key: None for key in string.digits}))
    arr = text.split()
    arr = [x for x in arr if len(x) > 1]
    return arr
```

```python
items = items.applymap(tokenize)
```

```python
clean_data = items.join(categories)
clean_data
```

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split

shuffled_data = sklearn.utils.shuffle(clean_data, random_state=0)
shuffled_data
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(shuffled_data['item_title'], shuffled_data['category'], random_state=0)
```
